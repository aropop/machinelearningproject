## The dataset

The data set for the assignment can be downloaded <a href='https://archive.ics.uci.edu/ml/machine-learning-databases/covtype/covtype.data.gz'>**here**</a>. You are asked to predict the forest cover type (the predominant kind of tree at a given location) from cartographic variables such as elevation, distance to water source, soil type, amount of shadow at certain hours of the day, etc. The actual forest cover type for a given 30 x 30 meter cell was determined from US Forest Service (USFS) Region 2 Resource Information System data. Independent variables were then derived from data obtained from the US Geological Survey and USFS. This study area includes four wilderness areas located in the Roosevelt National Forest of northern Colorado. These areas represent forests with minimal human-caused disturbances, so that existing forest cover types are more a result of ecological processes rather than forest management practices. The data is in raw form (not scaled) and contains binary columns of data for qualitative independent variables such as wilderness areas and soil type. A full description of the data set can be found <a href='https://archive.ics.uci.edu/ml/machine-learning-databases/covtype/covtype.info'>**here**</a>.

The data set can be loaded using following commands (make sure to put the dataset in your iPython notebook directory):

```python
import numpy as np
import pandas as pd

#read and randomly shuffle data
covtype = pd.read_csv('covtype.data').as_matrix()
covtype = covtype[np.random.permutation(covtype.shape[0]),:]

#assign train and test sets (in your experiments, you want to do cross-validation)
X_tr = covtype[0:40000,0:-1]
y_tr = covtype[0:40000,-1]
X_test = covtype[40001:-1,0:-2]
y_test = covtype[40001:-1,-1]
```

The following code projects the data to 2 dimensions to provide a visualization of the cover types. Each point corresponds to a patch of land with the color indicating the cover type (from blue =1 to red=7). 

```python
import pylab as pl
from sklearn.lda import LDA
%matplotlib inline

lda = LDA(n_components=2)
Xd = X_tr
X_r = lda.fit(Xd,y_tr).transform(Xd)
color_list = pl.cm.jet(np.linspace(0, 1, 7))

pl.figure()
for col, score in zip(color_list, range(1,8),):
    pl.scatter(X_r[y_tr == score, 0], X_r[y_tr == score, 1], c=col, label=repr(score))
pl.legend()
pl.show()
```

We can also look at individual features to get a feel for the data. For example, the first feature (column) in the dataset records the elevation of the piece of land that it describes. The code below provides a visualization of the differences in occurence of different types of trees depending on the elevation. You can see that Krummholz trees cluster at higher elevations, while Ponderosa Pine trees prefer lower elevations.

```python
pl.figure()
pl.hist(X_tr[y_tr==1,0],bins=10,color='r',alpha=0.5,label='Spruce/Fir')
pl.hist(X_tr[y_tr==2,0],bins=10,color='b',alpha=0.5,label='Lodgepole Pine')
pl.hist(X_tr[y_tr==3,0],bins=10,color='g',alpha=0.5,label='Ponderosa Pine')
pl.hist(X_tr[y_tr==7,0],bins=10,color='y',alpha=0.5,label='Krummholz')
pl.xlabel('elevation (m)')
pl.ylabel('#trees')
pl.title('#trees for given elevation')
pl.legend(loc=2)
pl.show()
```

# Solution

First we train a descion tree model. 

Before we do this we need to determine the ideal depth of the tree. This is done by splitting the data in a test group and a training group and comparing the error for each depth of tree.

```python
import numpy as np
import matplotlib.pylab as plt
from sklearn import tree
from sklearn import cross_validation
import numpy as np
import pandas as pd
import sys
%matplotlib inline

#read and randomly shuffle data
try:
  covtype
except NameError:
    covtype = pd.read_csv('covtype.data').as_matrix()
    covtype = covtype[np.random.permutation(covtype.shape[0]),:]


#assign train and test sets (in your experiments, you want to do cross-validation)
X_train = covtype[0:40000,0:-1] # Attributes
y_train = covtype[0:40000,-1] # Cover type
X_test = covtype[40001:-1,0:-1]
y_test = covtype[40001:-1,-1]

print("Imported data")

# Cross validation not really necessary as there are plenty of records

max_d = np.arange(1,25) # We calculate for depths 1-25
num_samples = np.arange(1,len(max_d) + 1)

train_score = np.zeros(len(max_d))
val_score = np.zeros(len(max_d))

optimal_model = 0

for idx,depth in enumerate(max_d):
    model = tree.DecisionTreeClassifier(max_depth=depth)
    model.fit(X_train,y_train)
    train_score[idx] = model.score(X_train,y_train)
    val_score[idx] = model.score(X_test,y_test)
    
    #show progress
    sys.stdout.write('\r')
    sys.stdout.write("Done processing depth " + str(depth) + " out of " + str(len(max_d)))
    sys.stdout.flush()

    
plt.figure()
plt.ylim((-0.1,0.5))

plt.plot(num_samples,1.0-train_score,label='training')
plt.plot(num_samples,1.0-val_score,'r',label='test')
plt.legend()


```

```python
#test the model for high bias or high variance
num_samples = np.arange(1,len(y_train), 200)

#tree depth
depth=18

n = len(num_samples)
#results on training/validation set
train_score = np.zeros(n)
val_score = np.zeros(n)

print(n)

for idx,samples in enumerate(num_samples):
    model = tree.DecisionTreeClassifier(max_depth=depth)
    model.fit(X_train[0:samples,:],y_train[0:samples])
    train_score[idx] = model.score(X_train[0:samples,:],y_train[0:samples])
    val_score[idx] = model.score(X_test,y_test)
    
    #show progress
    sys.stdout.write('\r')
    sys.stdout.write("Done processing idx " + str(idx))
    sys.stdout.flush()
    
#plot learning curves
plt.figure()
plt.ylim((-0.1,0.5))
#error rate = 1.0 - accurracy
plt.plot(num_samples,1.0-train_score,label='training')
plt.plot(num_samples,1.0-val_score,'r',label='validation')
plt.legend()
```

We can conclude that this model does not suffer from high variance or high bias as the error scores are relatively low and not very close together.

```python
import MultiLayerPerceptron as mlp
from sklearn import preprocessing


numhidden = [1,2,5,10,15,20]
restest = np.zeros(len(numhidden))
restrain = np.zeros(len(numhidden))

scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)

ind=0
# First find a optimal number of hidden units
for h in numhidden:
    net = mlp.MultiLayerPerceptron(54,h, 1, outfun=mlp.logistic, scorefun='ce') #input, hidden units, output
    net.fit(X_train_scaled,y_train)
    restest[ind] += net.score(X_test, y_test)
    restrain[ind] += net.score(X_train, y_train)
    ind +=1
    
    sys.stdout.write('\r')
    sys.stdout.write("Done processing hiddenunits " + str(h))
    sys.stdout.flush()

#plot performance
pl.figure()
pl.semilogy(numhidden, restest,'r',label='test') #use log scale for y axis
pl.semilogy(numhidden, restrain,label='train') #use log scale for y axis
pl.legend()
pl.show()
```
